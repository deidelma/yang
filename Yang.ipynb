# Yang

An implementation of the algorithm described by [Yang and Gorski](https://doi.org/10.1016/
j.xpro.2022.101283).

Currently only works on Intel Linux.  Did not install on Apple Silicon Linux because of failed compilation.

Modified the installation as follows:

Used conda to install python, jupyter, jupyterlab, R, r-essentials, r-tidyverse from conda-forge.
Used BiocManager to load the other packages.

```r
if(!requireNamespace("BiocManager", quietly=T)) install.packages("BiocManager")
```

```r
# note: we diverge from Yang and Gorski in that tidyverse already installed from conda-forge
required.packages=c("reshape2", "biomaRt", "org.Hs.eg.db", "edgeR", "limma", "NMF", "sva",
                   "vsn", "ggpubr", "ggfortify", "GSEABase")
```

```r
BiocManager::install(required.packages, ask=F)
```
